```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X, y, c="blue")
plt.show()
```

```python
def predict(X, theta):
    return X.dot(theta)

def cost(X, y, theta):
    m = len(y)
    predictions = predict(X, theta)
    cost = (1/(2*m)) * np.sum((predictions - y)**2)
    return cost

```

```python
X_b = np.c_[np.ones((100,1)), X]
theta = np.random.rand(2,1)
```

```python
cost(X_b, y, theta)
```

```python
theta
```

```python
costy = []
for i in range(100):
    theta[1] = i/10
    costy.append(cost(X_b, y, theta))
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(100), costy, c="blue")
plt.show()
```

```python
def gradient_descent(X, y, theta, learning_rate, iterations):
    costhist = np.zeros(iterations)
    m = len(y)
    for i  in range(iterations):
        gradients = (1/m) * X.T.dot(predict(X, theta) - y)
        theta = theta - learning_rate * gradients
        costhist[i] = cost(X, y, theta)
    
    return theta, costhist
```

```python
theta_new, cost_hist = gradient_descent(X_b, y, theta, 0.1, 100)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(100), cost_hist, c="blue")
plt.show()
```
